```python
dict = {'IDR': 1,
        'USD': 14479.35, 
        'GBP': 20020.81, 
        'SAR': 3860.66,
        'JPY': 131.43,
        'CNY': 2233.80}

in_curr = str(input("Pilih mata uang asal anda (IDR, USD, GBP, SAR, JPY, CNY) : "))
out_curr = str(input("Pilih mata uang tujuan konversi (IDR, USD, GBP, SAR, JPY, CNY) : "))
value = float(input("Masukkan jumlah mata uang yang ingin dikonversi: "))

def konvert (in_curr,out_curr,value):
  return (value/dict[in_curr])*dict[out_curr]

if value < 0 :
  print("Nilai jumlah mata uang tidak boleh negatif")
else:
  try:
    a = konvert(in_curr,out_curr,value)
    print("Nilai mata uang anda setelah konversi adalah",a,out_curr)
  except KeyError:
    print("Mata uang yang dikonversi tidak terdaftar atau salah")
```
